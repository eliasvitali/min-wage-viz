```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
from math import floor, ceil

#Min wage data: https://www.ncsl.org/research/labor-and-employment/state-minimum-wage-chart.aspx#Table

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

fPath = 'Data/counties/UScounties.shp'
map_data = gpd.read_file(fPath)

rent_data = pd.read_csv('Data/County_MedianRentalPrice_1Bedroom.csv')


#Remove Hawaii and Alaska
#map_data = map_data[~map_data['STATE_FIPS'].str.contains('02')]
#map_data = map_data[~map_data['STATE_FIPS'].str.contains('15')]

#Map only a single state
#map_data = map_data[map_data['STATE_FIPS'].str.contains('12')]

states = [f'{i:0>2}' for i in range(1, 57)]
omits = [f'{i:0>2}' for i in [3, 7, 14, 43, 52]]
for i in omits:
    states.remove(i)
    

rent_data['StateCodeFIPS']     = rent_data['StateCodeFIPS'].astype(str)
rent_data['MunicipalCodeFIPS'] = rent_data['MunicipalCodeFIPS'].astype(str)
rent_data['StateCodeFIPS']     = rent_data['StateCodeFIPS'].apply('{0:0>2}'.format)
rent_data['MunicipalCodeFIPS'] = rent_data['MunicipalCodeFIPS'].apply('{0:0>3}'.format)
rent_data['FIPS']              = rent_data['StateCodeFIPS'].map(str) + rent_data['MunicipalCodeFIPS']

rent_data = rent_data[['RegionName', 'FIPS', '2019-12']]

merged = map_data.set_index('FIPS').join(rent_data.set_index('FIPS'))

min_wage_data = pd.read_csv('Data/minwagedata.csv')
min_wage_data['FIPS'] = min_wage_data['FIPS'].apply('{0:0>2}'.format)

merged = merged.set_index('STATE_FIPS').join(min_wage_data.set_index('FIPS'))
merged['minwage'] = merged['minwage'].astype(float)

merged = merged.rename(index = str, columns = {
'2019-12': "rent"})

merged['hours'] = merged.apply(lambda row: row.rent/(4*row.minwage),
                axis=1)

merged = merged[['statename', 'NAME', 'rent', 'geometry', 'hours']]

### Plotting

comment = '''
fig, ax = plt.subplots(1, figsize = (20, 12) )
ax.axis('off')

for state in states[:5]:
    #merged = merged[merged['STATE_FIPS'].str.contains(state)]
    



#Plots 'No Data' layer
    merged.plot(ax = ax, 
                column = '2019-12', 
                color = 'grey', 
                label = 'No Data') 
    
    hours = list(merged['hours'])
    hours = [x for x in hours if str(x) != 'nan']
    
    
    hour_bins = range(floor(min(hours)), ceil(max(hours)), 4)
    hour_bins = [i for i in hour_bins]
    hour_bins
    
    
    #Plots data layer
    merged.dropna().plot(ax = ax, 
    #merged.plot(ax = ax, 
                         column = 'hours', 
                         cmap = 'viridis', 
                         scheme = 'User_Defined', 
                         #classification_kwds = dict( bins = [500, 800, 1000, 1200, 1500, 2000]),
                         classification_kwds = dict( bins = hour_bins), 
                         legend = True)
    plt.savefig(f'Images/{state}.png', dpi = 100)'''

merged.head()
```
